```python
import numpy as np
import pandas as pd
```

```python
data=pd.DataFrame({"a":[1,np.nan,3,4,5,6],"b":[8,9,np.nan,np.nan,3,4],"c":["q","w",np.nan,np.nan,np.nan,"y"],"d":[7,6,5,4,8,9]})
data
```

```python
data.isna()
```

```python
import matplotlib.pyplot as plt
```

```python
data.isna().sum().plot(kind="bar")
plt.show()
```

```python
data=data.dropna(subset=["a"])
data
```

```python
data["b"]=data["b"].fillna(data["b"].median())
data
```
